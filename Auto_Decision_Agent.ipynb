###рџ“„ STEP 2: Load PDF

```python
#рџ“„ STEP 2: Load your PDF
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("Python.pdf")
documents  = loader.load()
```

###вњ‚пёЏ STEP 3: Split PDF

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
  chunk_size = 350,chunk_overlap=50
 )

docs = splitter.split_documents(documents)
```

###рџ§  STEP 4: Embeddings

```python
!pip install sentence-transformers
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

###рџ“¦ STEP 5: Vector DB

```python
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(docs,embeddings)
```

### рџ”Ћ STEP 6: Add Web Search Tool

```python
from langchain_community.tools import DuckDuckGoSearchRun

search = DuckDuckGoSearchRun()
```

###рџ“„STEP 7:  PDF Tool

```python
def pdf_search(query):
    docs = db.similarity_search(query,k=2)
    return "\n".join([d.page_content for d in docs])
```

### Step 8 : рџ”ў Calculator Tool

```python
def calculator(query):
  try:
    return str(eval(query))
  except:
    return "Error in Calculation"
```

###рџ¤– STEP 9: LLM

```python
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline

pipe = pipeline("text-generation",model="gpt2")
llm = HuggingFacePipeline(pipeline=pipe)
```

### рџ§  STEP 10: Create Agent (MAIN рџ”Ґ)

```python
from langchain.agents.agent import AgentExecutor
from langchain.agents import create_react_agent
from langchain import hub

prompt = hub.pull("hwchase17/react")
#this is prebuild resoning prompt
```

### Create Agent

```python
from langchain.agents import Tool

tools = [
    Tool(
        name = "PDF_Search",
        func = pdf_search,
        description = "Use this to answer questions about the document"
    ),
    Tool(
        name = "Calculator",
        func = calculator,
        description = "Use this to answer math questions"
    ),
    Tool(
        name = "Web_Search",
        func = search.run,
        description = "Use this to answer questions about current events or unknown information"
    )
]

```

### рџљЂ STEP 11: Initialize Agent

```python
agent = create_react_agent(
    llm=llm,
    tools = tools,
    prompt=prompt
)
```

### рџ§Є STEP 7: Agent Executor

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True
)
```

```python
agent_executor.invoke({"input": "25 * 6"})
```
